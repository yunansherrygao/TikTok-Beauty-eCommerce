```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# transform list hashtags into string

df_cleaned['hashtags_str'] = df_cleaned['hashtags'].apply(lambda x: ' '.join(eval(x)) if isinstance(x, str) else '')

# TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=500, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_cleaned['hashtags_str'])

# K-Means into 5 clusters
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
df_cleaned['cluster'] = kmeans.fit_predict(tfidf_matrix)

# Check clusters' key words to name them
print("Each cluster's top keywords:")
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = tfidf.get_feature_names_out()
for i in range(n_clusters):
    print(f"Cluster {i}: ", end='')
    for ind in order_centroids[i, :5]:
        print(f'{terms[ind]} ', end='')
    print()
```

```python
# ANOVA
groups = [df_cleaned[df_cleaned['cluster'] == i]['engagement_rate'] for i in range(n_clusters)]
f_stat, p_val = stats.f_oneway(*groups)

print(f"\nANOVA F-statistic: {f_stat:.4f}, P-value: {p_val:.4f}")

# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster', y='engagement_rate', data=df_cleaned, estimator='mean', errorbar=None)
plt.title('Efficiency (Engagement Rate) by Hashtag Cluster')
plt.show()

if p_val < 0.05:
    print("Conclusion: ROAS between different clusters are statistically different")
else:
    print("Conclusion: ROAS between different clusters are not statistically different")
```

### 1.1 Multi regression analysis between Ad Content and ROAS
Analyzing video length, like rate, comment rate, and num of hashtag's influence on ROAS

```python
import statsmodels.api as sm
import numpy as np
import re

# extract hashtags
def extract_hashtags(text):
    if not isinstance(text, str): return ""
    tags = re.findall(r'#(\w+)', text)
    return " ".join(tags)

df_features = df_Fastmoss_ad[df_Fastmoss_ad['views'] > 0].copy()

df_features['extracted_hashtags'] = df_features['description'].apply(extract_hashtags)
df_features['hashtag_count'] = df_features['extracted_hashtags'].apply(lambda x: len(x.split()))

# calculate engagement rate
df_features['like_rate'] = df_features['likes'] / df_features['views']
df_features['comment_rate'] = df_features['comments'] / df_features['views']

# filter out those ROAS is 0 or missing
df_reg = df_features[df_features['ROAS'] > 0][['ROAS', 'duration_seconds', 'like_rate', 'comment_rate', 'hashtag_count']].dropna()

X = df_reg[['duration_seconds', 'like_rate', 'comment_rate', 'hashtag_count']]
y = df_reg['ROAS']

# Beta Coefficients
X_scaled = (X - X.mean()) / X.std()
X_scaled = sm.add_constant(X_scaled)

model_multi = sm.OLS(y, X_scaled).fit()
print(model_multi.summary())

# Visualize
plt.figure(figsize=(10, 6))
coef_df = model_multi.params[1:].reset_index()
coef_df.columns = ['Feature', 'Coefficient']
sns.barplot(x='Coefficient', y='Feature', data=coef_df.sort_values('Coefficient'), palette='coolwarm', hue='Feature', legend=False)
plt.title('Impact of Content Features on ROAS (Standardized Coefficients)')
plt.axvline(0, color='black', linestyle='--')
plt.show()
```

Surprisingly, based on the current dataset, none of the content features we tested—video duration, like rate, comment rate, or hashtag count—are statistically significant predictors of ROAS (all p-values are well above 0.05).

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Extract hashtags
def extract_hashtags(text):
    if not isinstance(text, str): return ""
    tags = re.findall(r'#(\w+)', text)
    return " ".join(tags)

df_Fastmoss_ad['extracted_hashtags'] = df_Fastmoss_ad['description'].apply(extract_hashtags)

# TF-IDF
tfidf = TfidfVectorizer(max_features=500, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_Fastmoss_ad['extracted_hashtags'])

# K-Means
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
df_Fastmoss_ad['topic_cluster'] = kmeans.fit_predict(tfidf_matrix)

print("Top 5 keywords for each cluster:")
terms = tfidf.get_feature_names_out()
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
for i in range(n_clusters):
    top_terms = [terms[ind] for ind in order_centroids[i, :5]]
    print(f"Cluster {i}: {', '.join(top_terms)}")

# ANOVA
df_analysis = df_Fastmoss_ad[df_Fastmoss_ad['ROAS'] > 0].copy()
groups = [df_analysis[df_analysis['topic_cluster'] == i]['ROAS'] for i in range(n_clusters)]
f_stat, p_val = stats.f_oneway(*groups)

print(f"\nANOVA F-statistic: {f_stat:.4f}, P-value: {p_val:.4f}")

# Visualize
plt.figure(figsize=(12, 6))
sns.barplot(x='topic_cluster', y='ROAS', data=df_analysis, estimator='mean', errorbar=None)
plt.title('Average ROAS by Hashtag Cluster')
plt.ylabel('Mean ROAS')
plt.show()
```

```python
# Check samapel count of each Cluster
print("Count of videos in each cluster:")
print(df_analysis.groupby('topic_cluster')['ROAS'].count())

# Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='topic_cluster', y='ROAS', data=df_analysis)
plt.title('ROAS Distribution by Hashtag Cluster (with Outliers)')
plt.ylabel('ROAS')
plt.show()

# Check each clusters' ROAS descriptive statistics
print("\nDescriptive Statistics of Each cluster's ROAS:")
display(df_analysis.groupby('topic_cluster')['ROAS'].describe())
```

#### Brand Clustering and ROAS Analysis (Fastmoss Advertiser Data)
We will cluster brands based on their scale and investment metrics, then evaluate their advertising efficiency (ROAS).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. Prepare features for clustering
# Features: creator_count, video_count, est_ad_spend, ad_revenue
features = ['creator_count', 'video_count', 'est_ad_spend', 'ad_revenue']
df_brands = df_Fastmoss_advertiser.dropna(subset=features).copy()

# 2. Scale the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_brands[features])

# 3. K-Means Clustering (let's start with 4 clusters to represent different brand scales)
kmeans_brands = KMeans(n_clusters=4, random_state=42, n_init=10)
df_brands['brand_cluster'] = kmeans_brands.fit_predict(scaled_features)

# 4. Calculate ROAS at the brand level
# To avoid division by zero, we filter for spend > 0
df_brands = df_brands[df_brands['est_ad_spend'] > 0]
df_brands['brand_ROAS'] = df_brands['ad_revenue'] / df_brands['est_ad_spend']

# 5. Profile the Clusters
brand_cluster_summary = df_brands.groupby('brand_cluster').agg({
    'brand_id': 'count',
    'creator_count': 'mean',
    'video_count': 'mean',
    'est_ad_spend': 'mean',
    'ad_revenue': 'mean',
    'brand_ROAS': 'mean'
}).rename(columns={'brand_id': 'brand_count'}).reset_index()

print("Brand Clusters Summary:")
display(brand_cluster_summary)

# 6. Visualize ROAS by Cluster
plt.figure(figsize=(10, 6))
sns.barplot(x='brand_cluster', y='brand_ROAS', data=df_brands, estimator='mean', errorbar=None)
plt.title('Average ROAS by Brand Cluster')
plt.show()
```

### 1.3 Correlational analysis between Ad Duration and ROAS

```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# confirm the format of ad_date is datetime
df_Fastmoss_advertiser['first_ad_date'] = pd.to_datetime(df_Fastmoss_advertiser['first_ad_date'])
df_Fastmoss_advertiser['last_ad_date'] = pd.to_datetime(df_Fastmoss_advertiser['last_ad_date'])

# Calculate duration
df_Fastmoss_advertiser['ad_duration_days'] = (df_Fastmoss_advertiser['last_ad_date'] - df_Fastmoss_advertiser['first_ad_date']).dt.days

# Filter out Ads that duration <= or ROAS is 0
df_analysis_duration = df_Fastmoss_advertiser[(df_Fastmoss_advertiser['ad_duration_days'] >= 0) & (df_Fastmoss_advertiser['ad_revenue'] > 0)].copy()

# Regression
X = df_analysis_duration['ad_duration_days']
y = df_analysis_duration['ad_revenue']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())

# Visualize
plt.figure(figsize=(12, 8))
sns.scatterplot(x='ad_duration_days', y='ad_revenue', data=df_analysis_duration, alpha=0.6)
sns.regplot(x='ad_duration_days', y='ad_revenue', data=df_analysis_duration, scatter=False, color='red', line_kws={'label': 'Linear Regression'})
plt.title('Advertising Duration vs. Ad Revenue')
plt.xlabel('Advertising Duration (Days)')
plt.ylabel('Ad Revenue')
plt.legend()
plt.grid(True)
plt.show()
```

The analysis shows a statistically significant relationship (P < 0.001) between the number of days a brand advertises and their total ad revenue. On average, each additional day of advertising is associated with an increase of approximately $2,711 in revenue.

However, the R-squared value is quite low (0.023), meaning that advertising duration only explains about 2.3% of the variation in revenue. This suggests that other factors—such as product category, creative quality, or budget—play a much larger role in determining total sales.


#### Regression with different brand clusters

```python
import statsmodels.api as sm

df_brands['first_ad_date'] = pd.to_datetime(df_brands['first_ad_date'])
df_brands['last_ad_date'] = pd.to_datetime(df_brands['last_ad_date'])
df_brands['ad_duration_days'] = (df_brands['last_ad_date'] - df_brands['first_ad_date']).dt.days

cluster_results = []

# Regression for every cluster
for cluster_id in sorted(df_brands['brand_cluster'].unique()):
    subset = df_brands[df_brands['brand_cluster'] == cluster_id]


    if len(subset) > 5:
        X = sm.add_constant(subset['ad_duration_days'])
        y = subset['ad_revenue']
        model = sm.OLS(y, X).fit()

        cluster_results.append({
            'Cluster': cluster_id,
            'Sample Size': len(subset),
            'Coefficient (Daily Increase)': model.params['ad_duration_days'],
            'P-value': model.pvalues['ad_duration_days'],
            'R-squared': model.rsquared
        })

results_df = pd.DataFrame(cluster_results)
display(results_df)

# Visualize
plt.figure(figsize=(14, 7))
sns.lmplot(x='ad_duration_days', y='ad_revenue', hue='brand_cluster', data=df_brands, aspect=1.5, scatter_kws={'alpha':0.4})
plt.title('Ad Duration vs Revenue by Brand Cluster')
plt.show()
```

Cluster 0 (Small/Average Scale): Shows a highly significant relationship (P < 0.0001). For these brands, each day of advertising adds about $451 in revenue, explaining 8.6% of their revenue variance.
Clusters 2 & 3 (Larger Scale): The P-values are much higher (0.29 and 0.61), meaning that for larger brands, simply increasing the number of days doesn't guarantee a linear rise in revenue.
This suggests that for smaller brands, 'persistence' is a key driver, while for larger brands, other strategic factors (like specific campaign quality or massive viral hits) are likely more influential than duration alone.

```python
import emoji
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extract emojis
def get_emoji_count(text):
    if not isinstance(text, str): return 0
    return emoji.emoji_count(text)

def extract_emojis(text):
    if not isinstance(text, str): return ""
    return ''.join(c for c in text if emoji.is_emoji(c))

df_emoji = df_Fastmoss_ad.copy()
df_emoji['emoji_count'] = df_emoji['description'].apply(get_emoji_count)
df_emoji['emojis'] = df_emoji['description'].apply(extract_emojis)

# Analysis the relationship between Emoji count and ROAS
plt.figure(figsize=(10, 6))
sns.scatterplot(x='emoji_count', y='ROAS', data=df_emoji[df_emoji['ROAS']>0], alpha=0.5)
plt.title('Emoji Count vs. ROAS')
plt.xlabel('Number of Emojis in Description')
plt.ylabel('ROAS')
plt.show()

correlation = df_emoji[df_emoji['ROAS']>0][['emoji_count', 'ROAS']].corr().iloc[0,1]
print(f"Correlation between Emoji count and ROAS is: {correlation:.4f}")
```
