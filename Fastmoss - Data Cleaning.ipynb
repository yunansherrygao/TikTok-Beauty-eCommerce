### Standardizing 'duration' and change them into integers in `ad_videos`/`medicube_videos_boosted`/`medicube_videos_organic`


```python
# Function to convert 'Xs' duration to seconds for ad_videos
def convert_ad_videos_duration(duration_str):
    if isinstance(duration_str, str) and duration_str.endswith('s'):
        return int(duration_str[:-1])
    return None # Handle unexpected formats or NaNs

# Apply the conversion to ad_videos
ad_videos['duration_seconds'] = ad_videos['duration'].apply(convert_ad_videos_duration)

# Display the first few rows with the new column and check data type for ad_videos
print("First 5 rows of `ad_videos` with new 'duration_seconds' column:")
display(ad_videos[['duration', 'duration_seconds']].head())
print("\nData type of 'duration_seconds' column in `ad_videos`:", ad_videos['duration_seconds'].dtype)
```

```python
# Function to convert 'HH:MM:SS' duration to seconds for medicube videos
def convert_hhmmss_duration(duration_str):
    if isinstance(duration_str, str) and len(duration_str.split(':')) == 3:
        h, m, s = map(int, duration_str.split(':'))
        return h * 3600 + m * 60 + s
    return None # Handle unexpected formats or NaNs

# Apply the conversion to medicube_videos_boosted using 'video_duration'
medicube_videos_boosted['duration_seconds'] = medicube_videos_boosted['video_duration'].apply(convert_hhmmss_duration)

# Display the first few rows and check data type for medicube_videos_boosted
print("First 5 rows of `medicube_videos_boosted` with new 'duration_seconds' column:")
display(medicube_videos_boosted[['video_duration', 'duration_seconds']].head())
print("\nData type of 'duration_seconds' in `medicube_videos_boosted`:", medicube_videos_boosted['duration_seconds'].dtype)
```

```python
# Apply the conversion to medicube_videos_organic using 'video_duration'
medicube_videos_organic['duration_seconds'] = medicube_videos_organic['video_duration'].apply(convert_hhmmss_duration)

# Display the first few rows and check data type for medicube_videos_organic
print("First 5 rows of `medicube_videos_organic` with new 'duration_seconds' column:")
display(medicube_videos_organic[['video_duration', 'duration_seconds']].head())
print("\nData type of 'duration_seconds' in `medicube_videos_organic`:", medicube_videos_organic['duration_seconds'].dtype)
```

```python
# Check for any remaining null values in 'duration_seconds' after conversion
print("Missing values in 'duration_seconds' of `ad_videos`:")
print(ad_videos['duration_seconds'].isnull().sum())

print("\nMissing values in 'duration_seconds' of `medicube_videos_boosted`:")
print(medicube_videos_boosted['duration_seconds'].isnull().sum())

print("\nMissing values in 'duration_seconds' of `medicube_videos_organic`:")
print(medicube_videos_organic['duration_seconds'].isnull().sum())
```
